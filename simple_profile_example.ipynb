```python
#!/usr/bin/env python3.7
####################################################
#
# Author: M Joyce
# For use with ASTR3007/6007 MESA-based assignments
#
####################################################
import numpy as np
import matplotlib.pyplot as plt

ZAMS_MESA_file = 'profile_files/near_ZAMS.profile'
RGBB_MESA_file = 'profile_files/near_RGBB.profile'
### add your last file here

#
# inspect your file to find the column indices of each quantity
#
logR, logRho = np.loadtxt(ZAMS_MESA_file, usecols=(2,4), unpack=True, skiprows=6)
logR2, logRho2 = np.loadtxt(RGBB_MESA_file, usecols=(2,4), unpack=True, skiprows=6)


plt.plot(logR, logRho, "go", markersize=3, label='ZAMS density Profile (log scale)')
plt.plot(logR2, logRho2, "ks", color='orange', markersize=3, label='RGBB density profile (log scale)')
plt.xlabel(r"logR [log($R/R_{\odot}$)]")
plt.ylabel(r"log $\rho$ [(g/cm$^3$)]")
plt.title("This plotting style is better b/c it accommodates more orders of magnitude")
plt.legend(loc=3, fontsize=12)
plt.show()
plt.close()


R = 10.0**logR
Rho = 10.0**logRho

plt.plot(R, Rho, "m*", markersize=3, label='Density Profile (linear scale)')
plt.xlabel("Radius [$R_{\odot}$]")
plt.ylabel(r"Density [(g/cm$^3$)]")
plt.legend(loc=1, fontsize=12)
plt.show()
plt.close()
```
