```python
#!/usr/bin/env python3.7
####################################################
#
# Author: M Joyce
# For use with ASTR3007/6007 MESA-based assignments
#
####################################################
import numpy as np
import matplotlib.pyplot as plt

MESA_file = 'trimmed_history.data'

#0 			  1 	   2 		 3 	    4		5
#model_number star_age star_mass log_L  log_R   log_Teff 
#
# inspect your file to find the column indices of each quantity
#
star_age, log_R= np.loadtxt(MESA_file, usecols=(1,4), unpack=True, skiprows=6)

star_age = star_age/1e9  ## converting age from years to Gyr

plt.plot(star_age, log_R, "mo", markersize=3,label='Radius vs Time')
plt.xlabel("Age (Gyr)")
plt.ylabel("Log R")
plt.legend(loc=2, fontsize=14)
plt.show()
```
